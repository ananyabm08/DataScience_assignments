```python
#	Detect and treat outliers if necessary.
def outlier_capping(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1  # Interquartile Range
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Apply capping
    df[column] = df[column].apply(lambda x: lower_bound if x < lower_bound else upper_bound if x > upper_bound else x)

# Loop to apply the function to all numeric columns
for col in df.select_dtypes(['int', 'float']).columns:
    outlier_capping(df, col)
```

# Task 2: Dimensionality Reduction with PCA:

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
```

```python
target = df['Type']           # Target variable y
features = df.drop('Type', axis=1)  # Feature variables x
```

##### 1.Standardize the features to ensure they have a mean of 0 and a standard deviation of Implement PCA to reduce the dimensionality of the dataset.

```python
scaler = StandardScaler()
scaled_df = scaler.fit_transform(features)
```

##### 2.Determine the optimal number of principal components using techniques like scree plot or cumulative explained variance.

```python
# Apply PCA
pca = PCA()
x_pca = pca.fit_transform(scaled_df)
x_pca
```

```python
# Plot cumulative explained variance
plt.figure(figsize=(10, 6))
plt.plot(np.cumsum(pca.explained_variance_ratio_), marker='o', linestyle='--', color='b')
plt.title('Cumulative Explained Variance by PCA Components')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid()
plt.show()
```

##### 3.Transform the original dataset into the principal components.

```python
#Keeping 3 principal components
pca = PCA(n_components=3)
pca
```

```python
x_pca_transformed = pca.fit_transform(scaled_df)
```

```python
# Convert to DataFrame for better visualization
pca_df = pd.DataFrame(data=x_pca_transformed, columns=['PC1', 'PC2', 'PC3'])
pca_df['Type'] = target.values  # Adding back the target for reference
```

```python
pca_df.head()
```

# Task 3: Clustering with Original Data:

##### 1. Apply a clustering algorithm (e.g., K-means) to the original dataset.

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score
import warnings
warnings.filterwarnings(action='ignore')
```

```python
# Apply K Means with 3 clusters (assumption based on domain or elbow method)
k=3
kmeans_original = KMeans(n_clusters=k,random_state=42)
kmeans_original
```

```python
original_cluster_labels = kmeans_original.fit_predict(scaled_df)
original_cluster_labels
```

```python
df["Cluster_Original"] = original_cluster_labels
df
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=features.iloc[:, 0], y=features.iloc[:, 1], hue=df["Cluster_Original"], palette="Set1")
plt.title("Clustering on Original Data (First Two Features)")
plt.show()
```

```python
# Evaluate clustering performance
sil_score_original = silhouette_score(scaled_df, original_cluster_labels)
db_score_original = davies_bouldin_score(scaled_df, original_cluster_labels)
```

```python
sil_score_original
```

```python
db_score_original
```

```python
print(f"Silhouette Score (Original Data): {sil_score_original:.3f}")
print(f"Davies-Bouldin Index (Original Data): {db_score_original:.3f}")
```

# Task 4: Clustering with PCA Data:

```python
kmeans_pca = KMeans(n_clusters=k,random_state=42)
kmeans_pca
```

```python
pca_cluster_labels = kmeans_pca.fit_predict(pca_df)
pca_cluster_labels
```

```python
# Silhouette Score
silhouette_pca = silhouette_score(pca_df, pca_cluster_labels)
db_pca = davies_bouldin_score(pca_df, pca_cluster_labels)
print(f"Silhouette Score (PCA Data): {silhouette_pca:.3f}")
print(f"DB Score (PCA Data): {db_pca:.3f}")
```

##### 2.Visualize the clustering results obtained from PCA-transformed data.

```python
# Plot clusters in PCA space
plt.figure(figsize=(10, 6))
sns.scatterplot(x=pca_df['PC1'], y=pca_df['PC2'], hue=pca_cluster_labels, palette='Set2', s=70)
plt.title('K-Means Clustering on PCA-Reduced Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()
```

##### 3.Compare the clustering results from PCA-transformed data with those from the original dataset.

```python
from sklearn.metrics import adjusted_rand_score

ari_score = adjusted_rand_score(original_cluster_labels, pca_cluster_labels)
print(f"Adjusted Rand Index (Original vs PCA Clustering): {ari_score:.3f}")
```

# Task 5: Comparison and Analysis:

```python
print("\nComparison of Clustering Results:")

print(f"Original Data - Silhouette Score: {sil_score_original:.3f}, Davies-Bouldin Index: {db_score_original:.3f}")
print(f"PCA Data - Silhouette Score: {silhouette_pca:.3f}, Davies-Bouldin Index: {db_pca:.3f}")
```
