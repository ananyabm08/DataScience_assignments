## 	Calculate the mean, median, mode, and standard deviation 

```python
# To select only numerical columns
num_col = df.select_dtypes(include = ['number'])
num_col.head()
```

### Calculate Mean

```python
mean = num_col.mean()
mean
```

### Calculate Median

```python
median = num_col.median()
median
```

### Calculate Mode

```python
mode = num_col.mode().iloc[0]
mode
```

### standard deviation

```python
std_dev = num_col.std()
std_dev
```

#### Summary

```python
summary_stats = pd.DataFrame({
    'Mean': mean,
    'Mode': mode,
    'Median': median,
    'Standard Deviation': std_dev
})
print(summary_stats)
```

# Data Visualization

## Histogram

```python
num_col.hist(bins=15, figsize=(12, 8), color='yellow', edgecolor='black')
plt.tight_layout()
plt.show()
```

## Boxplot

```python
sns.boxplot(data=num_col)
plt.title('Boxplot for all numerical columns', color='red', size=15, fontweight='bold')
plt.xticks(rotation=45)
plt.show()
```

```python
# Boxplot for one-by-one for numerical columns
for col in num_col.columns:
    plt.figure(figsize=(8,5))
    sns.boxplot(x=df[col], color='orange')

    plt.title(f'Boxplot of {col}', color='red', size=15, fontweight='bold')
    plt.xlabel(col, color='red', size=15)
    plt.grid(True)
    plt.show()
```

## Bar Chart

```python
cat_col = df.select_dtypes(include=['object'])
cat_col.head()
```

```python
# Bar chart for categorical columns
for col in cat_col.columns:
    plt.figure(figsize=(8,5))
    df[col].value_counts().plot(kind='bar', color='green')
    
    plt.title(f'Bar Chart for {col}', fontweight='bold', color='indigo', fontsize=15)
    plt.xlabel(col, fontsize=13, color='green')
    plt.ylabel('Count', fontsize=13, color='green')
    plt.xticks(rotation=45)
    plt.grid(axis='y')
    plt.show()
```

# Standardization of Numerical Variables

### before Standardization - Histogram

```python
#Before Standardization
num_col.hist(bins=20, figsize=(15, 8), color='pink')
plt.suptitle('Before Standardization',color='forestgreen',fontsize=18,fontweight='bold')
plt.show()
```

### Standardize

```python
from sklearn.preprocessing import StandardScaler
```

```python
# Create a scaler
std_sca = StandardScaler()

# Apply standardization
standardized_data = std_sca.fit_transform(num_col)

# Convert back to DataFrame for easy viewing
df_standardized = pd.DataFrame(standardized_data, columns=num_col.columns)

df_standardized.head()
```

### After standardization - Histogram

```python
#After Standardization
df_standardized.hist(bins=20, figsize=(15, 8), color='pink')
plt.suptitle('After Standardization',color='forestgreen',fontsize=18,fontweight='bold')
plt.show()
```

### Compare Before & After

```python
print("Original Mean:\n", num_col.mean())
print("\nStandardized Mean:\n", df_standardized.mean())

print("\nOriginal Std Dev:\n", num_col.std())
print("\nStandardized Std Dev:\n", df_standardized.std())
```
